# Atrapar porcentajes quiniela

```python
import bs4 as bs
import urllib
import urllib.request
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from random import randint
import time
import string
import pprint

import io
from urllib.request import urlopen, Request
import requests

import http.client
from http.client import IncompleteRead

from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import Select
```

```python
URL = "https://www.eduardolosilla.es/quiniela/ayudas/porcentajes"

MATCHES_                         = 14
MATCHES_MULT_LOOP_               = 5

PORCENTAJES_QUINIELISTA_ID_      = 0
PORCENTAJES_LAE_ID_              = 1
PORCENTAJES_REAL_ID_             = 2
PORCENTAJES_BASE_                = 4
PORCENTAJES_INCREMENT_           = 3

# download chromedriver last verion and put in this folder
driver = webdriver.Chrome('chromedriver')

driver.get(URL)

firstTime = True

jornada = 0
# for every "jornada"
while jornada < 58:
    time.sleep(randint(2,3))
    
    # Click on the "jornada" tab
    select = Select(driver.find_element_by_id('u-hidden-scrollbar'))
    select.select_by_value(str(jornada) )
    jornada += 1
    
    time.sleep(randint(10,15))    
    
    #results = driver.find_elements_by_xpath("//div[@class='c-caja_base m-no-fixed-size']")
    #print(results)
    #for result in results:
    #    print(result.text)
      
    #driver.find_elements_by_xpath("//div[@class='c-ayudas-porcentajes__parte-fija']")
    cajas = driver.find_elements_by_xpath("//div[@class='c-caja_base m-no-fixed-size']")
    
    porcentajes = driver.find_elements_by_xpath("//div[@class='c-porcentajes ng-star-inserted']")
    porcentajeList = []
    for porcentaje in porcentajes:
        porcentajeTable = porcentaje.text.split('\n')
        porcentajeList.append(porcentajeTable)
    #print(porcentajeList)
    
    
    for caja in cajas:
        cajaTable = caja.text.split('\n')
        #porcentajesTable = porcentajes[].text.split('\n')
        
        #print(caja.text)
        #print(cajaTable)

        iincrement5 = 0
        iincrement1 = 0
        while(iincrement5 < (MATCHES_MULT_LOOP_*MATCHES_)):
            
            line_ = cajaTable[iincrement5]
            #print(line_)
            saux = (cajaTable[iincrement5+1]).split(' - ')
            localTeamName_ = saux[0]
            #print(localTeamName_)
            awayTeamName_ = saux[1]
            #print(awayTeamName_)
            
            porcentajesQuiniela1_ = porcentajeList[PORCENTAJES_QUINIELISTA_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1)]
            porcentajesQuinielaX_ = porcentajeList[PORCENTAJES_QUINIELISTA_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1) + 1]
            porcentajesQuiniela2_ = porcentajeList[PORCENTAJES_QUINIELISTA_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1) + 2]
            #print(porcentajesQuiniela1_)
            #print(porcentajesQuinielaX_)
            #print(porcentajesQuiniela2_)
            porcentajesLAE1_ = porcentajeList[PORCENTAJES_LAE_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1)]
            porcentajesLAEX_ = porcentajeList[PORCENTAJES_LAE_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1) + 1]
            porcentajesLAE2_ = porcentajeList[PORCENTAJES_LAE_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1) + 2]
            
            porcentajesReal1_ = porcentajeList[PORCENTAJES_REAL_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1)]
            porcentajesRealX_ = porcentajeList[PORCENTAJES_REAL_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1) + 1]
            porcentajesReal2_ = porcentajeList[PORCENTAJES_REAL_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1) + 2]
            
            
            df = pd.DataFrame(columns=['Season', 'jornada', 'line', 'localTeamName', 'awayTeamName', 
                                       '1_porcentajesQuiniela', 'X_porcentajesQuiniela', '2_porcentajesQuiniela',
                                       '1_porcentajesLAE',      'X_porcentajesLAE',      '2_porcentajesLAE',
                                       '1_porcentajesReal',     'X_porcentajesReal',     '2_porcentajesReal'])  

            row = {'Season': '21-22',
            'jornada': jornada,
            'line': line_,
            'localTeamName': localTeamName_,
            'awayTeamName': awayTeamName_,
            '1_porcentajesQuiniela': porcentajesQuiniela1_,
            'X_porcentajesQuiniela': porcentajesQuinielaX_,
            '2_porcentajesQuiniela': porcentajesQuiniela2_,
            '1_porcentajesLAE': porcentajesLAE1_,
            'X_porcentajesLAE': porcentajesLAEX_,
            '2_porcentajesLAE': porcentajesLAE2_,
            '1_porcentajesReal': porcentajesReal1_,
            'X_porcentajesReal': porcentajesRealX_,
            '2_porcentajesReal': porcentajesReal2_
            }

            df = df.append(row, ignore_index=True )

            if firstTime:
                df_ant = df
                firstTime = False 
            else: 
                df_ant = pd.concat([df, df_ant], ignore_index=True)
            iincrement5 += 5
            iincrement1 += 1
            #print(df_ant)
            
            
            
    #holac-ayudas-porcentajes__parte-fija    
    
    
    #print(df_ant)
```
