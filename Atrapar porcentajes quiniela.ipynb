# Atrapar porcentajes quiniela

```python
import bs4 as bs
import urllib
import urllib.request
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from random import randint
import time
import string
import pprint

import io
from urllib.request import urlopen, Request
import requests

import http.client
from http.client import IncompleteRead

from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import Select
```

```python
URL = "https://www.eduardolosilla.es/quiniela/ayudas/"
TEMPORADA_ = 1 # 0 -> La ultima temporada

MATCHES_                         = 14
MATCHES_MULT_LOOP_               = 5

PORCENTAJES_QUINIELISTA_ID_      = 0
PORCENTAJES_LAE_ID_              = 1
PORCENTAJES_REAL_ID_             = 2
PORCENTAJES_BASE_                = 4
PORCENTAJES_INCREMENT_           = 3

# download chromedriver last verion and put in this folder
driverPorcentajes = webdriver.Chrome('chromedriver')
driverPorcentajes.get(URL + "porcentajes")

driverEscrutinio = webdriver.Chrome('chromedriver')
driverEscrutinio.get(URL + "escrutinio")


firstTime = True

time.sleep(randint(5,7))

#temporadaTab = driverPorcentajes.find_element_by_xpath("//*[starts-with(@class,'c-selector u-custom-select temporada')]/select")
#print(temporadaTab.text)
#temporadaList = temporadaTab.text.split('\n')
#for element in temporadaTab:
#    print(element.text)
select = Select(driverPorcentajes.find_element_by_xpath("//*[starts-with(@class,'c-selector u-custom-select temporada')]/select"))
select.select_by_value( str(TEMPORADA_) )

time.sleep(randint(4,6))

select = Select(driverEscrutinio.find_element_by_xpath("//*[starts-with(@class,'c-selector u-custom-select temporada')]/select"))
select.select_by_value(str(TEMPORADA_) )


jornada = 0
# for every "jornada"
while jornada < 58:
    time.sleep(randint(2,3))
    
    # Click on the "jornada" tab
    select = Select(driverPorcentajes.find_element_by_id('u-hidden-scrollbar'))
    select.select_by_value(str(jornada) )
    
    time.sleep(randint(2,3))
    
    select = Select(driverEscrutinio.find_element_by_id('u-hidden-scrollbar'))
    select.select_by_value(str(jornada) )
    
    jornada += 1
    
    time.sleep(randint(10,15))    
    
    #driver.find_elements_by_xpath("//div[@class='c-ayudas-porcentajes__parte-fija']")
    cajas = driverPorcentajes.find_elements_by_xpath("//div[@class='c-caja_base m-no-fixed-size']")
    
    porcentajes = driverPorcentajes.find_elements_by_xpath("//div[@class='c-porcentajes ng-star-inserted']")
    porcentajeList = []
    for porcentaje in porcentajes:
        porcentajeTable = porcentaje.text.split('\n')
        porcentajeList.append(porcentajeTable)
    #print(porcentajeList)
    
    #cajasResultados = driverEscrutinio.find_elements_by_xpath("//div[@class='c-caja_base m-no-fixed-size']/div/div/div]")
    #cajasResultados = driverEscrutinio.find_elements_by_xpath("//*[contains{@class,'u-pronostico m-resultado'}]")
    cajasResultados = driverEscrutinio.find_elements_by_xpath("//*[starts-with(@class,'u-pronostico m-resultado')]")
    resultadosList = []
    for element in cajasResultados:
        #resultadosTable = element.text.split('\n')
        resultadosList.append(element.text)
    #print(resultadosList)
        
    #c-tabla-categorias__table
    cajasRecaudacion = driverEscrutinio.find_elements_by_xpath("//table[@class='c-tabla-categorias__table']")
    recaudacionList = []
    for result in cajasRecaudacion:
        recaudacionTable = result.text.split(' ')
        recaudacionList.append(recaudacionTable)
    #print(recaudacionList)
    
    
    for caja in cajas:
        cajaTable = caja.text.split('\n')
        #porcentajesTable = porcentajes[].text.split('\n')
        
        #print(caja.text)
        #print(cajaTable)

        iincrement5 = 0
        iincrement1 = 0
        while(iincrement5 < (MATCHES_MULT_LOOP_*MATCHES_)):
            
            line_ = cajaTable[iincrement5]
            #print(line_)
            saux = (cajaTable[iincrement5+1]).split(' - ')
            localTeamName_ = saux[0]
            #print(localTeamName_)
            awayTeamName_ = saux[1]
            #print(awayTeamName_)
            
            result_ = resultadosList[iincrement1]
            
            porcentajesQuiniela1_ = porcentajeList[PORCENTAJES_QUINIELISTA_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1)]
            porcentajesQuinielaX_ = porcentajeList[PORCENTAJES_QUINIELISTA_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1) + 1]
            porcentajesQuiniela2_ = porcentajeList[PORCENTAJES_QUINIELISTA_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1) + 2]
            #print(porcentajesQuiniela1_)
            #print(porcentajesQuinielaX_)
            #print(porcentajesQuiniela2_)
            porcentajesLAE1_ = porcentajeList[PORCENTAJES_LAE_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1)]
            porcentajesLAEX_ = porcentajeList[PORCENTAJES_LAE_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1) + 1]
            porcentajesLAE2_ = porcentajeList[PORCENTAJES_LAE_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1) + 2]
            
            porcentajesReal1_ = porcentajeList[PORCENTAJES_REAL_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1)]
            porcentajesRealX_ = porcentajeList[PORCENTAJES_REAL_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1) + 1]
            porcentajesReal2_ = porcentajeList[PORCENTAJES_REAL_ID_][PORCENTAJES_BASE_ + (PORCENTAJES_INCREMENT_*iincrement1) + 2]
            
            acertantesAciertos15_ = recaudacionList[0][3]
            premioAciertos15_     = int("".join(filter(str.isdigit, recaudacionList[0][4][0:(recaudacionList[0][4]).find(',')])))
            acertantesAciertos14_ = recaudacionList[0][6]
            premioAciertos14_     = int("".join(filter(str.isdigit, recaudacionList[0][7][0:(recaudacionList[0][7]).find(',')])))
            acertantesAciertos13_ = recaudacionList[0][9]
            premioAciertos13_     = int("".join(filter(str.isdigit, recaudacionList[0][10][0:(recaudacionList[0][10]).find(',')])))
            acertantesAciertos12_ = recaudacionList[0][12]
            premioAciertos12_     = int("".join(filter(str.isdigit, recaudacionList[0][13][0:(recaudacionList[0][13]).find(',')])))
            acertantesAciertos11_ = recaudacionList[0][15]
            premioAciertos11_     = int("".join(filter(str.isdigit, recaudacionList[0][16][0:(recaudacionList[0][16]).find(',')])))
            acertantesAciertos10_ = recaudacionList[0][18]
            premioAciertos10_     = int("".join(filter(str.isdigit, recaudacionList[0][19][0:(recaudacionList[0][19]).find(',')])))
            acertantesElige8_     = recaudacionList[0][27]
            premioElige8_         = int("".join(filter(str.isdigit, recaudacionList[0][28][0:(recaudacionList[0][28]).find(',')])))
    
            
            df = pd.DataFrame(columns=['Season', 'jornada', 'line', 'localTeamName', 'awayTeamName', 'resultado', 
                                       '1_porcentajesQuiniela', 'X_porcentajesQuiniela', '2_porcentajesQuiniela',
                                       '1_porcentajesLAE',      'X_porcentajesLAE',      '2_porcentajesLAE',
                                       '1_porcentajesReal',     'X_porcentajesReal',     '2_porcentajesReal',
                                       'acertantes_15', 'premio_15', 'acertantes_14', 'premio_14',
                                       'acertantes_13', 'premio_13', 'acertantes_12', 'premio_12',
                                       'acertantes_11', 'premio_11', 'acertantes_10', 'premio_10',
                                       'acertantes_elige8', 'premio_elige8'
                                      ])  

            row = {'Season': '21-22',
            'jornada': jornada,
            'line': line_,
            'localTeamName': localTeamName_,
            'awayTeamName': awayTeamName_,
            'resultado': result_,
            '1_porcentajesQuiniela': porcentajesQuiniela1_,
            'X_porcentajesQuiniela': porcentajesQuinielaX_,
            '2_porcentajesQuiniela': porcentajesQuiniela2_,
            '1_porcentajesLAE': porcentajesLAE1_,
            'X_porcentajesLAE': porcentajesLAEX_,
            '2_porcentajesLAE': porcentajesLAE2_,
            '1_porcentajesReal': porcentajesReal1_,
            'X_porcentajesReal': porcentajesRealX_,
            '2_porcentajesReal': porcentajesReal2_,
            'acertantes_15':acertantesAciertos15_, 
            'premio_15':premioAciertos15_, 
            'acertantes_14':acertantesAciertos14_, 
            'premio_14':premioAciertos14_,                         
            'acertantes_13':acertantesAciertos13_, 
            'premio_13':premioAciertos13_,
            'acertantes_12':acertantesAciertos12_, 
            'premio_12':premioAciertos12_,
            'acertantes_11':acertantesAciertos11_, 
            'premio_11':premioAciertos11_,
            'acertantes_10':acertantesAciertos10_, 
            'premio_10':premioAciertos10_,
            'acertantes_elige8':acertantesElige8_, 
            'premio_elige8': premioElige8_  
            }

            df = df.append(row, ignore_index=True )

            if firstTime:
                df_ant = df
                firstTime = False 
            else: 
                df_ant = pd.concat([df_ant, df], ignore_index=True)
            iincrement5 += 5
            iincrement1 += 1
            #print(df_ant)
            
            
            
            
            
    #holac-ayudas-porcentajes__parte-fija    
    
    
    #print(df_ant)
```
